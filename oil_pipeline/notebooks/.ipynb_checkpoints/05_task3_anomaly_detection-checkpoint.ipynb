```python
# notebooks/05_task3_anomaly_detection.ipynb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
import boto3
from botocore.client import Config
import io
from datetime import timedelta

MINIO_CONFIG = {'endpoint': 'minio:9000', 'access_key': 'minioadmin', 'secret_key': 'minioadmin123', 'secure': False, 'bucket': 'oil-data-lake'}
s3 = boto3.client('s3', endpoint_url=f"http://{MINIO_CONFIG['endpoint']}", aws_access_key_id=MINIO_CONFIG['access_key'], aws_secret_access_key=MINIO_CONFIG['secret_key'], config=Config(signature_version='s3v4'))

def read(key):
    resp = s3.get_object(Bucket=MINIO_CONFIG['bucket'], Key=key)
    return pd.read_parquet(io.BytesIO(resp['Body'].read()))

pumps = read('raw/pumps.parquet')
sensors = read('raw/pump_sensors.parquet')
failures = read('raw/pump_failures.parquet')

sensor_cols = ['temperature','vibration','current','rpm','pressure']
scaler = StandardScaler()
scaled = scaler.fit_transform(sensors[sensor_cols].fillna(sensors[sensor_cols].median()))

iso_forest = IsolationForest(contamination=0.1, random_state=42)
sensors['anomaly'] = iso_forest.fit_predict(scaled)
sensors['is_anomaly'] = sensors['anomaly'] == -1

print(f"Anomalies found: {sensors['is_anomaly'].sum()}")

risk_features = ['temperature','vibration','current','rpm','pressure']
positive = []
for _, f in failures.iterrows():
    ftime = pd.to_datetime(f['failure_date'])
    for h in [1,3,6,12]:
        sample = sensors[(sensors['pump_id']==f['pump_id']) & (pd.to_datetime(sensors['timestamp'])==ftime - timedelta(hours=h))]
        if not sample.empty:
            s = sample.iloc[0].to_dict()
            s['risk'] = 1
            positive.append(s)

normal = sensors[~sensors['is_anomaly']].sample(min(len(positive)*3, len(sensors)))
normal['risk'] = 0

all_samples = pd.concat([pd.DataFrame(positive), normal], ignore_index=True)
X_risk = all_samples[risk_features].fillna(all_samples[risk_features].median())
y_risk = all_samples['risk']

scaler_risk = StandardScaler()
X_scaled = scaler_risk.fit_transform(X_risk)

rf_risk = RandomForestClassifier(n_estimators=100, random_state=42)
rf_risk.fit(X_scaled, y_risk)

last = sensors.groupby('pump_id').last().reset_index()
X_last = scaler_risk.transform(last[risk_features].fillna(last[risk_features].median()))
risk_scores = rf_risk.predict_proba(X_last)[:,1]

risk_df = pd.DataFrame({'pump_id': last['pump_id'], 'risk_score': risk_scores})
risk_df['risk_level'] = pd.cut(risk_scores, bins=[0,0.3,0.7,1], labels=['Low','Medium','High'])
risk_df = risk_df.merge(pumps[['pump_id','well_id','type']], on='pump_id')

buf = io.BytesIO()
risk_df.to_parquet(buf, index=False)
buf.seek(0)
s3.put_object(Bucket=MINIO_CONFIG['bucket'], Key='predictions/risk_assessment.parquet', Body=buf.getvalue())

print(risk_df)
```
