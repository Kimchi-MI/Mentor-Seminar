```python
# notebooks/01_etl_to_minio.ipynb
import pandas as pd
from sqlalchemy import create_engine
import boto3
from botocore.client import Config
from io import BytesIO

DB_CONFIG = {'host': 'postgres', 'port': 5432, 'database': 'oil_db', 'user': 'oil_user', 'password': 'oil_pass'}
engine = create_engine(f"postgresql://{DB_CONFIG['user']}:{DB_CONFIG['password']}@{DB_CONFIG['host']}:{DB_CONFIG['port']}/{DB_CONFIG['database']}")

MINIO_CONFIG = {'endpoint': 'minio:9000', 'access_key': 'minioadmin', 'secret_key': 'minioadmin123', 'secure': False, 'bucket': 'oil-data-lake'}
s3 = boto3.client('s3', endpoint_url=f"http://{MINIO_CONFIG['endpoint']}", aws_access_key_id=MINIO_CONFIG['access_key'], aws_secret_access_key=MINIO_CONFIG['secret_key'], config=Config(signature_version='s3v4'))

try:
    s3.create_bucket(Bucket=MINIO_CONFIG['bucket'])
except:
    pass

tables = ['wells','production','well_telemetry','well_targets','pumps','pump_sensors','pump_failures','drivers','vehicles','deliveries','oil_stations']

for table in tables:
    df = pd.read_sql(f"SELECT * FROM {table}", engine)
    buffer = BytesIO()
    df.to_parquet(buffer, index=False)
    buffer.seek(0)
    s3.put_object(Bucket=MINIO_CONFIG['bucket'], Key=f"raw/{table}.parquet", Body=buffer.getvalue())
```
