```python
# notebooks/02_data_preparation.ipynb
import pandas as pd
import numpy as np
import boto3
from botocore.client import Config
import io

MINIO_CONFIG = {'endpoint': 'minio:9000', 'access_key': 'minioadmin', 'secret_key': 'minioadmin123', 'secure': False, 'bucket': 'oil-data-lake'}
s3 = boto3.client('s3', endpoint_url=f"http://{MINIO_CONFIG['endpoint']}", aws_access_key_id=MINIO_CONFIG['access_key'], aws_secret_access_key=MINIO_CONFIG['secret_key'], config=Config(signature_version='s3v4'))

def read(key):
    resp = s3.get_object(Bucket=MINIO_CONFIG['bucket'], Key=key)
    return pd.read_parquet(io.BytesIO(resp['Body'].read()))

wells = read('raw/wells.parquet')
production = read('raw/production.parquet')
telemetry = read('raw/well_telemetry.parquet')
targets = read('raw/well_targets.parquet')

production['temperature'] = production['temperature'].fillna(production.groupby('well_id')['temperature'].transform('median')).fillna(production['temperature'].median())
production['pressure'] = production['pressure'].fillna(production.groupby('well_id')['pressure'].transform('median')).fillna(production['pressure'].median())

telemetry['date'] = pd.to_datetime(telemetry['timestamp']).dt.date
daily_telemetry = telemetry.groupby(['well_id', 'date']).agg({'pump_speed_rpm': 'mean', 'pump_current': 'mean', 'pressure_in': 'mean', 'pressure_out': 'mean', 'temperature': 'mean', 'vibration': 'mean', 'oil_flow_rate': 'mean'}).reset_index()
daily_telemetry.columns = ['well_id', 'date', 'avg_pump_speed_rpm', 'avg_pump_current', 'avg_pressure_in', 'avg_pressure_out', 'avg_temperature', 'avg_vibration', 'avg_oil_flow_rate']

production['downtime_ratio'] = production['downtime_hours'] / 24
production['efficiency'] = production['oil_ton'] / (production['energy_kwh'] + 1)

ml_data = production.merge(daily_telemetry, on=['well_id','date'], how='left').merge(targets, on=['well_id','date'], how='left').merge(wells[['well_id','region','operator','status']], on='well_id', how='left')

daily_production = production.groupby('date').agg({'oil_ton':'sum','gas_m3':'sum','water_m3':'sum','energy_kwh':'sum','downtime_hours':'mean'}).reset_index()
well_kpi = production.groupby('well_id').agg({'oil_ton':['mean','sum'],'downtime_ratio':'mean','efficiency':'mean'}).round(2)
well_kpi.columns = ['avg_oil_ton','total_oil_ton','avg_downtime_ratio','avg_efficiency']
well_kpi = well_kpi.reset_index().merge(wells[['well_id','name']], on='well_id')

def save(df, key):
    buf = io.BytesIO()
    df.to_parquet(buf, index=False)
    buf.seek(0)
    s3.put_object(Bucket=MINIO_CONFIG['bucket'], Key=key, Body=buf.getvalue())

save(production, 'processed/production_clean.parquet')
save(daily_telemetry, 'processed/daily_telemetry.parquet')
save(ml_data, 'processed/ml_dataset.parquet')
save(daily_production, 'marts/daily_production.parquet')
save(well_kpi, 'marts/well_kpi.parquet')
```
