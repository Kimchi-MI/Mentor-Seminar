```python
# notebooks/03_task1_production_analytics.ipynb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import boto3
from botocore.client import Config
import io

MINIO_CONFIG = {'endpoint': 'minio:9000', 'access_key': 'minioadmin', 'secret_key': 'minioadmin123', 'secure': False, 'bucket': 'oil-data-lake'}
s3 = boto3.client('s3', endpoint_url=f"http://{MINIO_CONFIG['endpoint']}", aws_access_key_id=MINIO_CONFIG['access_key'], aws_secret_access_key=MINIO_CONFIG['secret_key'], config=Config(signature_version='s3v4'))

def read(key):
    resp = s3.get_object(Bucket=MINIO_CONFIG['bucket'], Key=key)
    return pd.read_parquet(io.BytesIO(resp['Body'].read()))

production = read('processed/production_clean.parquet')
wells = read('raw/wells.parquet')
well_kpi = read('marts/well_kpi.parquet')
daily_production = read('marts/daily_production.parquet')

plt.figure(figsize=(12,5))
plt.plot(daily_production['date'], daily_production['oil_ton'], marker='o')
plt.title('Daily Oil Production')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('/home/jovyan/work/charts/daily_production.png')

plt.figure(figsize=(10,6))
well_prod = production.groupby('well_id')['oil_ton'].sum().reset_index().merge(wells[['well_id','name']], on='well_id')
plt.barh(well_prod['name'], well_prod['oil_ton'])
plt.title('Total Production by Well')
plt.tight_layout()
plt.savefig('/home/jovyan/work/charts/well_production.png')

corr = production[['oil_ton','temperature','pressure','energy_kwh','downtime_hours']].corr()['oil_ton']
print(corr)

plt.figure(figsize=(10,5))
plt.scatter(production['temperature'], production['oil_ton'], c=production['pressure'], cmap='coolwarm', alpha=0.6)
plt.colorbar(label='Pressure')
plt.xlabel('Temperature')
plt.ylabel('Oil Production')
plt.savefig('/home/jovyan/work/charts/temp_pressure_scatter.png')
```
